# data import

```python
import pandas as pd
```

```python
raw_data = pd.read_csv("https://raw.githubusercontent.com/connoralydon/simplify-job-scraper/main/output.csv")
raw_data.head(10)
```

## adding and modifying attributes

```python
# add remote
raw_data["remote"] = raw_data["title"].str.contains("remote") | raw_data["location"].str.contains("remote")
```

```python
# dropping columns that don't add value
# my collection method made these values unusable
raw_data.drop(columns=["archived","screen","interview","applied"], inplace=True, errors='ignore')
```

```python
# converting to bool
for col in ["offer","rejected"]:
    raw_data[col] = raw_data[col].astype(int)
    raw_data[col] = raw_data[col].astype(bool)
```

```python
# taking out special whitespace characters
raw_data["title"] = raw_data["title"].str.replace("\s"," ",regex=True)
raw_data["location"] = raw_data["location"].str.replace("\s"," ",regex=True)
raw_data["location"] = raw_data["location"].str.replace("usa"," ")
```

```python
# converting dates to date time
raw_data["date_opened"] = pd.to_datetime(raw_data["date_opened"])
```

```python
raw_data.to_csv("jobs.csv", index = False)
```
